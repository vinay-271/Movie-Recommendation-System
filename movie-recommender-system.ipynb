```python
import pandas as pd 
import numpy as np
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.info()
```

```python
credits.info()
```

```python
# joining movies and credits based on Title

movies = movies.merge(credits,on="title")
movies.info()
```

```python
# listing useful feaures for recommendation system 
# id is important for fetching movies from database
useful = ["genres","id","keywords","title","overview","cast","crew"]

movies = movies[useful]
movies.info()
```

```python
movies.isnull().sum()
```

```python
# only 3 movies are there that doesn't have Overview, therefore dropping them wont be an issue
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies["genres"].iloc[0]
```

```python
# we need to change it from 
# [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]
# ["Action","Adventure","Fantasy","Science Fiction"]

# We create a helper function called Convert 
"""
it translates the string obj into a real list so the for loop can extract the "name" values, doing so securely without the security risks of the standard eval() function.
"""
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
movies["genres"] = movies["genres"].apply(convert)
```

```python
movies.head()
```

```python
# now we'll apply convert on Keywords and other similar columns 
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i["name"])
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies["crew"] = movies["crew"].apply(fetch_director)
```

```python
# converting overview feature values into a llist

movies["overview"] = movies["overview"].apply(lambda x:x.split())
```

### Removing Spaces between words

```python
movies["genres"] = movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"] = movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"] = movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
# Now we'll create a new feature tags which is combination of Overviwe, genres, keywords, cast and crew 
movies["tags"] = movies["overview"] + movies["genres"] + movies["keywords"] + movies["cast"] + movies["crew"]
```

```python
# dropping all other columns
new_movies = movies[["id","title","tags"]]
```

```python
new_movies
```

```python
new_movies["tags"] = new_movies["tags"].apply(lambda x:" ".join(x))
```

```python
# converting content of tags to lowercase
new_movies["tags"] = new_movies["tags"].apply(lambda x:x.lower())
```

```python
new_movies.head()
```

# Vectorization

```python
# applying stemming 
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_movies["tags"] = new_movies["tags"].apply(stem)
```

```python
# text to vectors (text vectorization)

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000 , stop_words='english')
```

```python
vectors = cv.fit_transform(new_movies["tags"]).toarray()
```

```python
cv.get_feature_names_out()
```

```python
# now we'll calculate the cosine distance between the vectors 
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

# Main Function for Recommendation

```python
def recommend(movie):
    movie_idx = new_movies[new_movies["title"]==movie].index[0]
    distances = similarity[movie_idx]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_movies.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
